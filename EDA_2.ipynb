```python
import numpy as np
import pandas as pd
import os

from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
?WordCloud
```

```python
# Get all the fake news text files into a list -- FAKE
FAKE = []
#use full texts (not normalized, truncated) for testing difference
dir1 = '/Users/braulio/Documents/Projects/FakeNewsD/Data/fake/'
list_files = os.listdir(dir1)
number_files = len(list_files)
```

```python
for i in range(number_files):
        title = list_files[i]
        
        with open(dir1 + title,'r') as reader:
            doc = reader.read()
            doc.lower()
            doc.split()
            reader.close
            FAKE.append(doc)
```

```python
len(FAKE)
```

```python
# Start with one review:
text = FAKE[0]
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)
```

```python
# Display the generated image:
plt.figure(figsize=(14,6))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text = " ".join(FAKE)
print ("There are {} words in the combination of all FAKE news.".format(len(text)))
```

```python
stopwords_br = ['como', 'um', 'que', 'para', 'como', 'uma', 'de', 'ou', 'os', 'em', 'por', 'ma',
                 'se', 'era', 'ela', 'ele', 'era', 'seu', 'sua', 'mai', 'mas', 'nos', 'isso', 'mas', 'entre', 
                 'entre', 'outro', 'na', 'até', 'sobre', 'essa', 'esse', 'quando', 'mais', 'ao', 'dos', 'tem', 
                  'dos', 'está', 'não', 'da', 'disse', 'diz', 'foi', 'caso', 'outra', 'também', 'já', 'ano', 
                'ainda', 'são', 'anos', 'sem', 'porque', 'há', 'eles', 'segundo', 'afirmou', 'acordo', 'muito', 
               'todo', 'muito', 'conta', 'mesmo', 'dele', 'todos', 'dia', 'depois', 'então', 'tudo', 'seja', 'quem'
               'toda', 'após', 'só', 'agora', 'hoje', 'foram', 'grupo', 'assim', 'fazer', 'desde', 'eu']
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(stopwords_br)
```

```python
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
```

```python
# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(14,6))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
